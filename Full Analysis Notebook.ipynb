## Reading Data and joining to deal with them as one

```python
pd.set_option('display.max_columns',None)
products=pd.read_csv('products.csv')
orders=pd.read_csv('orders.csv')
orders_reviews=pd.read_csv('order_reviews.csv')
orders_payments=pd.read_csv('order_payments.csv')
orders_items=pd.read_csv('order_items.csv')
customers=pd.read_csv('customers.csv')
category_tranlation=pd.read_csv('category_translation.csv')
```

```python
#Joining Orders data
orders_final=pd.merge(pd.merge(orders,orders_items,on='order_id'),orders_payments,on='order_id').merge(orders_reviews,on='order_id')
```

```python
#Joining all data
za=pd.merge(pd.merge(orders_final,products,on='product_id'),customers,on='customer_id').merge(category_tranlation,on='product_category_name')
```

```python
za.isna().sum()
za.drop('review_comment_title',axis=1,inplace=True) ## contatin more than 100k missing values
```

```python
#Checking data types
za.dtypes.to_frame()[:10]
```

```python
za['order_purchase_timestamp']=pd.to_datetime(za['order_purchase_timestamp'])
za['order_approved_at']=pd.to_datetime(za['order_approved_at'])
za['order_delivered_carrier_date']=pd.to_datetime(za['order_delivered_carrier_date'])
za['order_delivered_customer_date']=pd.to_datetime(za['order_delivered_customer_date'])
za['order_estimated_delivery_date']=pd.to_datetime(za['order_estimated_delivery_date'])
```

```python
def period(hours):
    if hours in range(6,13):
        return ('Morning')
    elif hours in range(13,19):
        return ('Afternoon')
    else:
        return ('Evening')  
```

#### Extracting more details for Purchasing time

```python
za['order_purchase_month']=za['order_purchase_timestamp'].dt.month
za['order_purchase_year']=za['order_purchase_timestamp'].dt.year
za['order_purchase_day']=za['order_purchase_timestamp'].dt.day
za['order_purchase_day_name']=za['order_purchase_timestamp'].dt.day_name()
za['order_purchase_hour']=za['order_purchase_timestamp'].dt.hour
za['order_purchase_period']=za['order_purchase_hour'].apply(period)
```

### Approval time

#### Difference between Order Approval and order Purchasing

```python
za['purchasing_approval']=za['order_approved_at']-za['order_purchase_timestamp']
za['purchasing_approval']=za['purchasing_approval'].dt.days
```

```python
review_approval=za.groupby('purchasing_approval').agg({'review_score':'mean'}).sort_values(by='purchasing_approval',ascending=False)[:10].rename(columns={'purchasing_approval':'Days to be Approved','review_score':'Review Score'})
review_approval
```
