## Applying RFM analysis and KMeans for clustering

What is RFM (recency, frequency, monetary) analysis? RFM analysis is a marketing technique used to quantitatively rank and group customers based on the recency, frequency and monetary total of their recent transactions to identify the best customers and perform targeted marketing campaigns. The system assigns each customer numerical scores based on these factors to provide an objective analysis. RFM analysis is based on the marketing adage that "80% of your business comes from 20% of your customers."

RFM analysis ranks each customer on the following factors:

Recency. How recent was the customer's last purchase? Customers who recently made a purchase will still have the product on their mind and are more likely to purchase or use the product again. Businesses often measure recency in days. But, depending on the product, they may measure it in years, weeks or even hours.

Frequency. How often did this customer make a purchase in a given period? Customers who purchased once are often are more likely to purchase again. Additionally, first time customers may be good targets for follow-up advertising to convert them into more frequent customers.

Monetary. How much money did the customer spend in a given period? Customers who spend a lot of money are more likely to spend money in the future and have a high value to a business.

What is a good RFM score? The best RFM score is the one with the highest values for each variable. If a store uses a 1 to 5 scale for recency, frequency, and monetary, with 5 being the highest, then the perfect RFM score is 555.

```python
za.head()
```

```python
max_phurchase_date = max(za['order_purchase_timestamp']).date()
max_phurchase_date
```

```python
rfm_table = za.groupby('customer_unique_id').agg({'order_purchase_timestamp': lambda x:(datetime.strptime(str(max_phurchase_date),'%Y-%m-%d')-x.max()).days,'order_id':'count','price':'sum'})                                                                      
rfm_table.rename(columns={'order_purchase_timestamp':'Recency','order_id':'Frequency','price':'Monetary'},inplace=True)
```

```python
rfm_table_copy=rfm_table.copy()
```

```python
rfm_table['R']  = pd.qcut(rfm_table['Recency'], 5, labels=[5, 4, 3, 2, 1])
rfm_table['F']= pd.qcut(rfm_table['Frequency'].rank(method="first"), 5, labels=[1, 2, 3, 4, 5])
rfm_table['M'] = pd.qcut(rfm_table['Monetary'], 5, labels=[1, 2, 3, 4, 5])
```

```python
def join_rfm(data):
    return str(data['R'])+str(data['F'])+str(data['M'])
rfm_table['RFM_Segment']=rfm_table.apply(join_rfm,axis=1)
rfm_table['RFM_Score']=rfm_table[['R','M','F']].sum(axis=1)
```

```python
def rfm_level(data):
    if data['RFM_Segment'] in(['555','554', '544', '545', '454', '455', '445']):
        return 'Champions'
    elif data['RFM_Segment']in(['543', '444', '435', '355', '354', '345', '344', '335']):
        return 'Loyal'
    elif data['RFM_Segment']in(['553', '551', '552', '541', '542', '533', '532', '531', '452', '451', '442', '441', '431', '453', '433', '432', '423', '353', '352', '351', '342', '341', '333', '323']):
        return 'Potential Loyalists'       
    elif data['RFM_Segment']in(['512', '511', '422', '421', '412', '411', '311']):
        return 'New Customers'        
    elif data['RFM_Segment']in(['525', '524', '523', '522', '521', '515', '514', '513', '425','424', '413','414','415', '315', '314', '313']):
        return 'Promising'   
    elif data['RFM_Segment']in(['535', '534', '443', '434', '343', '334','325', '324']):
        return 'Need Attention'   
    elif data['RFM_Segment']in(['331', '321', '312', '221', '213', '231', '241', '251']):
        return 'About To Sleep'                   
    elif data['RFM_Segment']in(['255', '254', '245', '244', '253', '252', '243', '242', '235', '234', '225', '224', '153', '152', '145', '143', '142', '135', '134', '133', '125', '124']):
        return 'At Risk'      
    elif data['RFM_Segment']in(['155', '154', '144', '214','215','115', '114', '113']):
        return 'Cannot Lose Them'    
    elif data['RFM_Segment']in(['332', '322', '233', '232', '223', '222', '132', '123', '122', '212', '211']):
        return 'Hibernating customers'              
    elif data['RFM_Segment']in(['111', '112', '121', '131','141','151']):
        return 'Lost customers'                       
```

```python
rfm_table['Segment']=rfm_table.apply(rfm_level,axis=1)
```

#### Applying Clustering for Customer segmentaion by K-means

```python
rfm_table_copy.reset_index(inplace=True)
```

```python
rfm_table_copy.drop('customer_unique_id',axis=1,inplace=True)
```

##### Scaling 

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
data_scaled=scaler.fit_transform(rfm_table_copy)
```

```python
data_scaled=pd.DataFrame(data_scaled,columns=scaler.get_feature_names_out())
data_scaled=data_scaled[['Frequency','Recency','Monetary']]
```

```python
data_scaled
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
wcss=[]
scores=[]
for i in range(2,15):
    kmeans=KMeans(n_clusters=i)
    y_kmeans=kmeans.fit_predict(data_scaled)
    wcss.append(kmeans.inertia_)
    scores.append(silhouette_score(data_scaled,y_kmeans))
plt.figure(1 , figsize = (12,6))
plt.plot(np.arange(2,15) , wcss , 'o')
plt.xlabel('Number of Clusters') , plt.ylabel('Inertia')
plt.plot(np.arange(2, 15) , wcss, alpha = 0.5)
plt.show()
```

```python
plt.figure(1 , figsize = (12,6))
plt.plot(np.arange(2,15) , scores , 'o')
plt.xlabel('Number of Clusters') , plt.ylabel('Sil Score')
plt.plot(np.arange(2, 15) , scores, alpha = 0.5)
plt.show()
```

#### Now we Can try 5 Clusters

```python
kmeans=KMeans(n_clusters=5,init='k-means++',random_state=42)
y_kmeans=kmeans.fit_predict(data_scaled)
```
